```python
# get set up library

import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import PIL 
from PIL import ImageEnhance

#import image
from google.colab.patches import cv2_imshow
img = cv2.imread('sample_data/Narenda_Wicaksono.jpg')

cv2_imshow(img)
```

```python
#convert color
fig = plt.figure()
fig.add_subplot(121)

plt.imshow(img)

fig.add_subplot(122)

img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img2)

plt.show()
```

```python
#brightness filter

img = PIL.Image.open('sample_data/Narenda_Wicaksono.jpg') 
converter = ImageEnhance.Brightness(img) 
img2 = converter.enhance(0.5) 
img3 = converter.enhance(2)

# Plot image 
plt.gcf().set_size_inches(15,15) 
plt.subplot(131),plt.imshow(img),plt.title('Original') 
plt.xticks([]), plt.yticks([]) 
plt .subplot(132),plt.imshow(img2),plt.title('Half Brightness') 
plt.xticks([]), plt.yticks([]) 
plt.subplot(133),plt.imshow(img3),plt.title('Double Brightness') 
plt.xticks([]), plt.yticks([]) 

plt.imshow(img)
plt.imshow(img2)
plt.imshow(img3)
```

```python
#grayscale image

img = cv2.imread('sample_data/Narenda_Wicaksono.jpg',0)
img = cv2.medianBlur(img,5)

ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,\
            cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#flip image
img = cv2.imread('sample_data/Narenda_Wicaksono.jpg')
print(type(img))

print(img.shape)

img_flip_ud = cv2.flip(img, 0)
cv2.imwrite('sample_data/Narenda_flip_ud.jpg', img_flip_ud)

img_flip_lr = cv2.flip(img, 1)
cv2.imwrite('sample_data/Narenda_flip_lr.jpg', img_flip_lr)
```
